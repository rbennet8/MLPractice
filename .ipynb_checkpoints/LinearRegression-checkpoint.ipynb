# Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston = datasets.load_boston()
x = boston.data
y = boston.target
z = boston.feature_names

bostonDF = pd.DataFrame(data=x, columns=z)
bostonDF.head()
```

```python
plt.figure(figsize=(15, 10))
plt.hist(y)
plt.xlabel('price ($1000s)')
plt.ylabel('count')
```

```python
for index, feature_name in enumerate(z):
    plt.figure(figsize=(10, 8))
    plt.scatter(x[:, index], y, s=10)
    plt.ylabel('Price', size=15)
    plt.xlabel(feature_name, size=15)
    plt.tight_layout()
```

```python
correlation_matrix = bostonDF.corr().round(2)
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
# Run linear regression on both the most correlated in the map above
# and the feature that's most correlated with price, from the scatter
# plots above that.

# Make sure to mention what the features are and talk about if they're important.
```

```python
X_train, X_test, y_train, y_test = train_test_split(np.array(bostonDF['TAX']).reshape(-1, 1), np.array(bostonDF['RAD']).reshape(-1, 1), test_size = 0.25)
```

```python
model = LinearRegression()

# Fit the model on the data
model.fit(X_train, y_train)

# Predictions using the trained model
y_pred = model.predict(X_test)
```

```python
# Plot the data points and the linear regression line
plt.scatter(X_train, y_train, label='Data')
plt.plot(X_test, y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
